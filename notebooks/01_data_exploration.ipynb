```python
# Imports
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# Load ratings.dat
ratings = pd.read_csv(
    "../data/raw/ml-1m/ratings.dat",
    sep="::",
    engine="python",
    names=["user_id", "movie_id", "rating", "timestamp"]
)

ratings.head()
```

```python
# Load movies
movies = pd.read_csv(
    "../data/raw/ml-1m/movies.dat",
    sep="::",
    engine="python",
    encoding="latin-1",
    names=["movie_id", "title", "genres"]
)

movies.head()
```

```python
# Load Users
users = pd.read_csv(
    "../data/raw/ml-1m/users.dat",
    sep="::",
    engine="python",
    names=[
        "user_id",
        "gender",
        "age",
        "occupation",
        "zip_code"
    ]
)

users.head()
```

```python
# Basic dataset statistics
print("Users:", ratings["user_id"].nunique())
print("Movies:", ratings["movie_id"].nunique())
print("Interactions:", len(ratings))
```

```python
# Sparsity
num_users = ratings["user_id"].nunique()
num_movies = ratings["movie_id"].nunique()
num_interactions = len(ratings)

possible_interactions = num_users * num_movies

print("Possible interactions:", possible_interactions)
print("Observed interactions:", num_interactions)

sparsity = 1 - (num_interactions / possible_interactions)

print(f"Sparsity: {sparsity * 100:.2f}%")
```

```python
# Rating distribution
ratings["rating"].value_counts().sort_index()
```

```python
# Timestamp range
ratings["datetime"] = pd.to_datetime(
    ratings["timestamp"],
    unit="s"
)

print("Earliest:", ratings["datetime"].min())
print("Latest:", ratings["datetime"].max())
```

```python
# Duplicate user-movie interactions
duplicates = ratings.duplicated(
    subset=["user_id", "movie_id"]
).sum()

print("Duplicate user-movie pairs:", duplicates)
```

```python
# User Activity
user_counts = ratings.groupby("user_id").size()

print(user_counts.describe())

movie_counts = ratings.groupby("movie_id").size()

print(movie_counts.describe())
```

```python
positive_ratings = ratings[ratings["rating"] >= 4]

print("Positive interactions:", len(positive_ratings))

print(
    f"Percentage retained: "
    f"{100 * len(positive_ratings) / len(ratings):.2f}%"
)
```
